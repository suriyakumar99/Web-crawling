```python
from urllib.request import urlopen
import re
from nltk.stem import WordNetLemmatizer

lem = WordNetLemmatizer()
def cleaninput(text):
    try:
        text = re.sub("[\",\\.-:#()\[\]?;'!{}]","",text)
        text = re.sub("[-_+|@~`*=<>]","",text)
        text = re.sub('\d+',"", text)
        text = text.split()
        for i,t in enumerate(text):
            text[i] = lem.lemmatize(t)
            text[i] = text[i].lower()
    except Exception as e:
        print(e)
    return text
```

```python
words = []
term_docid = []

urls = [
    "http://textfiles.com/science/genetic.txt",
    "http://textfiles.com/science/blackhol.txt",
    "http://textfiles.com/science/discover.txt",
    "http://textfiles.com/science/fea.txt",
    "http://textfiles.com/science/htt.txt",
    "http://textfiles.com/science/hazchem.txt",
    "http://textfiles.com/science/kennedy.txt",
    "http://textfiles.com/science/mars1.txt",
    "http://textfiles.com/science/monitors.txt",
    "http://textfiles.com/news/article.txt"
    ]
for i in range(len(urls)):
    text = urlopen(urls[i])
    text = str(text.read(),'utf-8')
    text = cleaninput(text)
    words.append([w.lower() for w in text])
    for word in text:
        term_docid.append((word.lower(),i+1))

term_docid = sorted(set(term_docid))
finalans = []
print ("word\t\t\tfrequency\t\tposting list\n")
for t in term_docid:
    a = t[0]
    b = 0
    c = ""
    for i in range(len(urls)):
        if a in words[i]:
            b+=1
            c = c+ str(i+1) + "->"
    finalans.append(a+"\t\t\t"+str(b)+"\t\t"+c[:-2])

finalans = sorted(set(finalans))

for x in finalans:
    print(x)
```
